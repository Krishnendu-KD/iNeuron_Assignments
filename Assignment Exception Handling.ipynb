**Q 1. Write a function to compute 5/0 and use try/except to catch the exceptions**

```python
numerator=int(input("Enter Numerator:"))
denominator=int(input("Enter Denominator:"))

try:
    result=numerator/denominator
    print(result)
    
except Exception as e:
    print("Exception {}, Try non-zero denominator".format(e))
```

```python
def handleException(subjects,verbs,objects):
    if ("Americans" not in subjects) or ("Indians" not in subjects):
        raise Exception("Americans or Indians should be entered in the subjects")
    elif ("play" not in verbs) or ("watch" not in verbs):
        raise Exception("play or watch should be entered in the verbs")
    elif ("Cricket" not in objects) or ("Baseball" not in objects):
        raise Exception("Baseball or Cricket should be entered in the verbs")
        
    else:
        for i in subjects:
            for j in verbs:
                for k in objects:
                    print("{} {} {}.".format(i,j,k))

subjects=input("Type subject names :").split(',')
verbs=input("Type the verbs :").split(',')
objects=input("Type object names :").split(',')

try:
    handleException(subjects, verbs, objects)
except Exception as e:
    print(e)
    
```
